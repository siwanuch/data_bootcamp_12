# **Order Pizza**

```python
## Order Pizza
def chat():
    menu = {"hawaiian":20, "cheese":30, "hotdog":10}
    bill_order = {"hawaiian":0, "cheese":0, "hotdog":0}
    bill = 0
    user_name = input("Hi May I have your name: ")
    print(f"Welcome {user_name} !")

    print(f"Our menus are: ")

    for key, value in menu.items():
        print(f"    {key}: ${value}")

    while True:
        order = input("What can I get for you?: ")
        qty = int(input("How many?: "))
        if order == 'hawaiian':
            bill_order["hawaiian"] = bill_order["hawaiian"] + qty
        elif order == 'cheese':
            bill_order["cheese"] = bill_order["cheese"] + qty
        elif order == 'hotdog':
            bill_order["hotdog"] = bill_order["hotdog"] + qty
        else:
            print("Your input is invalid")
            continue

        check = input("Anything else? (Y/N): ")
        if check=='N':
            # Use a dictionary comprehension to multiply values for shared keys
            total_costs = {
            item: bill_order[item] * menu[item]
            for item in bill_order
            if item in menu
            }

            for value in total_costs.values():
                bill = bill + value
            print(f"Your bill is $ {bill}. \nThank you!")
            break
        elif check == 'Y':
            continue
        else:
            print("Your input is invalid")
            continue
```
