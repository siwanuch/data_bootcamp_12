## **Game: Rock Paper Scissor**

```python
## Rock Paper Scissor
import random
def pyc():
    """
    To play Pao Ying Chup
    """
    score = 0
    hands = ("kon","gungai","kradad")
    count = 0
    while True:
        user_hand = input("Choose your hand (or Type 'exit'): ")

        if user_hand == 'exit':
            print(f"Your final score is: {score}. Good bye")
            break

        op_hand = random.choice(hands)
        print(f"Your opponent hand: {op_hand}")

        if user_hand == op_hand:
            score = score
        elif user_hand == "gungai":
            if op_hand == "kradad":
                score = score+1
            elif op_hand == "kon":
                score = score-1
        elif user_hand == "kradad":
            if op_hand == "kon":
                score = score+1
            elif op_hand == "kon":
                score = score-1
        elif user_hand == "kon":
            if op_hand == "gungai":
                score = score+1
            elif op_hand == "kradad":
                score = score-1
        else:
            print("Your input is invalid, choose again")
            continue
        print(f"Your score is: {score}")
```

```python
pyc()
```
