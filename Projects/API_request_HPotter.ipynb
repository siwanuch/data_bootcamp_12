```python
import requests
import pandas as pd
import time

url = "https://hp-api.onrender.com/api/characters"
response = requests.get(url)

data = response.json()

# Only grab name,house,ancestry from each item in the JSON list
filtered_data = [
    {'Name': item['name'], 'House': item['house'], 'Ancestry':['ancestry']}
    for item in data
]

# Convert list of dictionaries to DataFrame
df = pd.DataFrame(filtered_data)
print(df.head())
```
