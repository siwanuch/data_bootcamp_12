```python
import random

class ATM:
    def __init__(self, username, bank, amount):
        self.username = username
        self.bank = bank
        self.amount = amount
    def __str__(self):
        return f"{self.username} uses {self.bank} and has {self.amount} THB in an account."
    def check_balance(self):
        print(f"Balance: {self.amount} THB")
    def get_otp(self):
        otp = random.randint(10000,99999)
        return f"Your Requested OTP: {otp}"
    def deposit(self,saving):
        self.amount += saving
        print(f"Deposited {saving} THB successfully!")
    def withdraw(self,wdraw):
        self.amount -= wdraw
        print(f"Withdrew {wdraw} THB successfully!")
    def transfer(self,tranf,target_bank):
        self.amount -= tranf
        print(f"Transferred {tranf} THB to {target_bank} successfully")
    def paybill(self,bill,payee):
        print(f"Paid {bill} THB to {payee} successfully")
```

```python
tk = ATM("siwanuch","scb",200)
```
